```python
import csv
import sys
import pandas as pd
import os
import glob
from collections import Counter


path  = r"/soe/hlitz/maxwell/notebooks/Prefetching_SSD/Data/output/SYSTOR/part10" 
output  = r"/soe/hlitz/maxwell/notebooks/Prefetching_SSD/Data/output/SYSTOR/part10" 
names = ['TimeStamp', 'Response_Time','IO_Type','LUN','ByteOffset', 'IOSize']




all_files = glob.glob(os.path.join(path, "*.csv"))
print(len(all_files))




for file in all_files:
    spl  = file.split("/")
    p = len(spl)
    name = spl[p-1]
    print()
    print(name)
    with open(file) as f:
        df = pd.DataFrame(columns=names)
        TimeStamp = []
        Response_Time = []
        IO_Type = []
        LUN = []
        ByteOffset = []
        IOSize = []
        
        for line in f:
            line = str(line)
            k  = line.split(",")
            if(len(k)<5):
                continue
            if('ByteOffset' in k or "Offset" in k):
                continue
            TimeStamp.append(k[0].strip())
            Response_Time.append(k[1].strip())
            IO_Type.append(k[2].strip())
            LUN.append(k[3].strip())
            ByteOffset.append(k[4].strip())
            IOSize.append(k[5].strip())
            
            
    df['TimeStamp']=TimeStamp  
    df['Response_Time']= Response_Time
    df['IO_Type']= IO_Type
    df['LUN']=  LUN
    df['ByteOffset']=  ByteOffset
    df['IOSize']= IOSize
    

    
    filename_wr_out = os.path.join(output,name)
    df.to_csv(filename_wr_out,index=False)
    df = pd.read_csv(filename_wr_out,engine='python',index_col=False)
    
    
    x = Counter(df['ByteOffset'])
    vals = {}
    vals =  x.most_common(1000)
    bo_list = []
    coverage = 0
    for key,value in vals:
        bo_list.append(key)
        coverage = coverage + value
    

    print("Percentage Coverage")
    print((coverage/len(df))*100)
    
  
    
    df["ByteOffset"]= df["ByteOffset"].astype(float) 
    df['ByteOffset_Delta'] = df['ByteOffset'] - df['ByteOffset'].shift(-1)
    
    x = Counter(df['ByteOffset_Delta'])
    vals = {}
    vals =  x.most_common(1000)


    bo_list = []
    coverage = 0
    for key,value in vals:
        bo_list.append(key)
        coverage = coverage + value
    

    print("Percentage Coverage")
    print((coverage/len(df))*100)
    print(len(df))
```
