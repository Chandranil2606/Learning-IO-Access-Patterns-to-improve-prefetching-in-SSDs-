```python
# Author: Chandranil Chakraborttii
# Project: Learning I/O Access Patterns to Improve Prefetching in SSDs
# Paper Link :https://www.researchgate.net/profile/Chandranil_Chakraborttii/publication/344379801_Learning_IO_Access_Patterns_to_Improve_Prefetching_in_SSDs/links/5f6e28fba6fdcc00863adb13/Learning-I-O-Access-Patterns-to-Improve-Prefetching-in-SSDs.pdf
# Loading libraries

import csv
import sys
import pandas as pd
import os
import glob

```

```python


path  = r"/soe/hlitz/maxwell/notebooks/Prefetching_SSD/Data/output/Buildserver/" 
output  = r"/soe/hlitz/maxwell/notebooks/Prefetching_SSD/Data/output/Buildserver/filtered_data" 

#filename_rd = os.path.join(path,'A.trace.csv')
names = ['Operation','TimeStamp','Process_Name','ThreadID','IrpPtr','ByteOffset','IOSize','ThreadID','ElapsedTime','DiskNum','IrpFlags','DiskSvcTime','IOPri','VolSnap','FileObject','FileName','IO_Pri']



#filename_wr_2 = os.path.join(output,'old_op.csv')

all_files = glob.glob(os.path.join(path, "*.csv"))     # advisable to use os.path.join as this makes concatenation OS independent
print(all_files)
```

```python
# Load data
# Data Transformation


for file in all_files:
    spl  = file.split("/")
    p = len(spl)
    name = spl[p-1]
    print(name)
    with open(file) as f:
        df = pd.DataFrame(columns=names)
        Operation = [] #0
        TimeStamp = [] #1
        Process_Name = [] #2
        ThreadID = [] #3
        IrpPtr = [] #4
        ByteOffset = [] #5
        IOSize = [] #6
        ThreadID= [] #7
        ElapsedTime= [] #8
        DiskNum= [] #9
        IrpFlags= [] #10
        DiskSvcTime= [] #11
        IOPri= [] #12
        VolSnap= [] #13
        FileObject= [] #14 
        FileName= [] #15

        for line in f:
            line = str(line)
            k  = line.split(",")
            if(len(k)<17):
                continue
            if("TimeStamp" in k):
                continue
                
            # Handle int operations
            TimeStamp.append(int(k[1]))
            ThreadID.append(int(k[3]))
            ElapsedTime.append(int(k[8]))
            DiskNum.append(int(k[9]))
            DiskSvcTime.append(int(k[11]))
#             IOPri.append(k[16])
#             print(k[16])
            s= k[16].rstrip()
            IOPri.append(s)

            
            # Handle hex to int operations
            IrpPtr.append(int(k[4],16))
            ByteOffset.append(int(k[5],16))
            IOSize.append(int(k[6],16))
            IrpFlags.append(int(k[10],16))
            FileObject.append(int(k[14],16))

            
            # Handle enum operations (Save as string, handle later)
            Operation.append(k[0])
            Process_Name.append(k[2])
            VolSnap.append(k[13])
            FileName.append(k[15])
            
            
            
            
            
    df['Operation']=Operation  
    df['TimeStamp']= TimeStamp
    df['Process_Name']= Process_Name
    df['ThreadID']=ThreadID
    df['IrpPtr']=  IrpPtr
    df['ByteOffset']=  ByteOffset
    df['IOSize']= IOSize
    df['ElapsedTime']=ElapsedTime  
    df['DiskNum']= DiskNum
    df['IrpFlags']= IrpFlags
    df['DiskSvcTime']=DiskSvcTime
    df['IO_Pri']=  IOPri
    df['VolSnap']=  VolSnap
    df['FileObject']= FileObject
    df['FileName']= FileName
    
    
    a = df['Operation'].unique().tolist()
    thread_id_map = {}
    for i,id in enumerate(a): thread_id_map[id] = i 
    df['Operation_class'] = df['Operation'].map(lambda x: thread_id_map[x])

    a = df['Process_Name'].unique().tolist()
    thread_id_map = {}
    for i,id in enumerate(a): thread_id_map[id] = i 
    df['Process_Name_class'] = df['Process_Name'].map(lambda x: thread_id_map[x])


    a = df['VolSnap'].unique().tolist()
    thread_id_map = {}
    for i,id in enumerate(a): thread_id_map[id] = i 
    df['VolSnap_class'] = df['VolSnap'].map(lambda x: thread_id_map[x])


    a = df['FileName'].unique().tolist()
    thread_id_map = {}
    for i,id in enumerate(a): thread_id_map[id] = i 
    df['FileName_class'] = df['FileName'].map(lambda x: thread_id_map[x])


    filename_wr_out = os.path.join(output,name)
    df.to_csv(filename_wr_out,index=False)
```
